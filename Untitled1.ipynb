```python
import numpy as np
import matplotlib.pyplot as plt
import gglens_x_ggclustering_likelihood as utils
```

```python
config = utils.config('config_r_2_30_3_30_snweight_allcosmoparams_fiducialprior_rsd_meascorr_magcorr_dm_ldpz_NS.json', use_interp_meas_corr=False)
config_mag = utils.config('config_r_2_30_3_30_snweight_allcosmoparams_fiducialprior_rsd_meascorr_magcorr_dm_ldpz_NS_mag.json', use_interp_meas_corr=False)
config = utils.config('config_temp.json', use_interp_meas_corr=False)
#config_mag = utils.config('config_mock_fiducial_mag.json', use_interp_meas_corr=False)
```

```python
x,y,cov = config.load_data()
x,y,covmag = config_mag.load_data()
```

```python
c = cov.get_covariance('lensing')
cmag = covmag.get_covariance('lensing')
```

```python
r = np.zeros(c.shape)
for i in range(c.shape[0]):
    for j in range(c.shape[0]):
        r[i][j] = c[i][j] /(c[i][i]*c[j][j])**0.5
        
rmag = np.zeros(cmag.shape)
for i in range(cmag.shape[0]):
    for j in range(cmag.shape[0]):
        rmag[i][j] = cmag[i][j] /(cmag[i][i]*cmag[j][j])**0.5
```

```python
plt.figure(figsize=(6,6))
im = plt.imshow(r,cmap='Blues')
plt.colorbar(im)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
n = 3
nr = int(rmag.shape[0]/3)
samples = ['LOWZ', 'CMASS1', 'CMASS2']
im = plt.imshow(rmag,cmap='Blues')
cbar = plt.colorbar(im)
cbar.ax.set_ylabel(r'$r_{ij}^{\rm full}$')
ax_pos = ax.get_position()
cbar_pos0 = cbar.ax.get_position()
cbar_pos1 = [cbar_pos0.x0, ax_pos.y0, cbar_pos0.x1 - cbar_pos0.x0, ax_pos.y1 - ax_pos.y0]
cbar.ax.set_position(cbar_pos1)
for i in range(n):
    ax.axhline(i*nr-0.5, color='k')
    ax.axvline(i*nr-0.5, color='k')
plt.setp(ax.get_xticklabels(), visible=False)
plt.setp(ax.get_yticklabels(), visible=False)
ax.tick_params(axis='both', which='minor', length=0)
ax.tick_params(axis='both', which='major', length=0)
for i in range(n):
    ax.text((i+0.5)*nr, -1, samples[i], ha="center", va="bottom")
    ax.text(-1, (i+0.5)*nr, samples[i], ha="right", va="center", rotation=90)
for i in range(n):
    ax.text(0 +i*nr    , nr*n+1, r'$R_{\rm min}$', ha='left' , va='top',fontsize=15)
    ax.text(nr+i*nr-1.0, nr*n+1, r'$R_{\rm max}$', ha='right', va='top',fontsize=15)
```
